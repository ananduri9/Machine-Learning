```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
#load data
iris = load_iris()
print("Shape: ", iris.data.shape)

X = iris.data
Y = iris.target
```

```python
#plot first three dimensions of iris
x_plot = X[:, :3]
y_plot = Y

fig = plt.figure(figsize=(6,6))
plot = Axes3D(fig, elev=48, azim=133)
plot.scatter(x_plot[:,0], x_plot[:,1], x_plot[:,2],
            c=y_plot, edgecolor='k')
plot.set_xlabel('Petal Width')
plot.set_ylabel('Sepal length')
plot.set_zlabel('Petal Length')
plot.set_title('Iris Dataset')
```

```python
#fit KMeans clustering
n_clusters = 3
clust = KMeans(n_clusters=n_clusters)
clust.fit(X)
labels = clust.labels_
```

```python
#plot clustered

fig = plt.figure(figsize=(6,6))
plot = Axes3D(fig, elev=48, azim=133)
plot.scatter(x_plot[:,0], x_plot[:,1], x_plot[:,2],
            c=labels, edgecolor='k')
plot.set_xlabel('Petal Width')
plot.set_ylabel('Sepal length')
plot.set_zlabel('Petal Length')
plot.set_title('Clustered Iris Dataset (Clusters = {})'.format(n_clusters))
```
