```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
#load data
iris = load_iris()
print("Shape: ", iris.data.shape)

#split to train and test data
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, train_size=0.6, test_size=0.4)
```

```python
#plot first three dimensions of iris
x_plot = x_train[:, :3]
y_plot = y_train

fig = plt.figure(figsize=(6,6))
plot = Axes3D(fig, elev=48, azim=133)
plot.scatter(x_plot[:,0], x_plot[:,1], x_plot[:,2],
            c=y_plot, edgecolor='k')
plot.set_xlabel('Petal Width')
plot.set_ylabel('Sepal length')
plot.set_zlabel('Petal Length')
plot.set_title('Iris Dataset')
```

```python
#Build Support Vector Machine
SV_clf = SVC(kernel='rbf', gamma='auto', C=1.0)
SV_clf.fit(x_train, y_train)
SV_score = SV_clf.score(x_test, y_test)

print("The mean accuracy for SVM is: ", SV_score)
```

```python
#Build Adaboost
AB_clf = AdaBoostClassifier(n_estimators=150, learning_rate=1)
AB_clf.fit(x_train, y_train)
AB_score = AB_clf.score(x_test, y_test)

print("The mean accuracy for adaboost is: ", AB_score)
```
